```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
```

<h1>For Dataset1<h1/>

```python
dataset=pd.read_excel('Dataset1.xlsx')
dataset
```

```python
from mlxtend.frequent_patterns import apriori

frequent_itemset= apriori(dataset,min_support=0.1,use_colnames=True)
frequent_itemset
```

```python
from mlxtend.frequent_patterns import association_rules
result=association_rules(frequent_itemset,metric="confidence",min_threshold=0.3)
#Print Final Rules
result
```

```python
final_result=result[['antecedents','consequents','support','confidence','lift']]
final_result
```
